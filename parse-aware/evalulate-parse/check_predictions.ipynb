```python
import os

model_dir = "/shared/3/projects/national-culture/cache/independent/cache/input-to-parse/checkpoint-1875"
predictions_path = os.path.join(model_dir, 'i2p_all_predictions.txt')
references_path = os.path.join(model_dir, 'i2p_references.txt')

def load_file_lines(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = [line.strip() for line in file]
    return lines

predicted_sn = load_file_lines(predictions_path)
reference_en = load_file_lines(references_path)

from datasets import load_dataset

dataset = load_dataset("rahular/itihasa", split="test")
input_en = [example["translation"]["en"] for example in dataset.select(range(len(reference_en)))]
label_sn = [example["translation"]["sn"] for example in dataset.select(range(len(predicted_sn)))]

print("Alignment Check for First 10 Examples:\n")
for i in range(10):
    print(f"Example {i + 1}:")
    print(f"  Input English (input_en): {input_en[i]}")
    print(f"  Reference English (reference_en): {reference_en[i]}")
    print(f"  Label Sanskrit (label_sn): {label_sn[i]}")
    print(f"  Predicted Sanskrit (predicted_sn): {predicted_sn[i]}")
    print()
```
